```
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras import datasets
from tensorflow.keras import models,layers
```

(a). Fetching the training and test datasets for CIFAR10

```
(x_train, y_train), (x_test, y_test) = datasets.cifar10.load_data()
```

(b). Creating a validation set of 10000 images from the training set.

```
#can be done at the time of fiting the model using validation split.
```

(c). Training the CNN architectures as follows :

(d). **MODEL1**

```
model1 = models.Sequential()
model1.add(layers.Conv2D(32,(5,5),activation='relu',input_shape=(32,32,3)))
model1.add(layers.MaxPool2D((2,2)))
model1.add(layers.Conv2D(64,(5,5),activation='relu'))
model1.add(layers.MaxPool2D((2,2)))
model1.add(layers.Flatten())
model1.add(layers.Dense(64,activation='relu'))
model1.add(layers.Dense(10,activation='softmax'))
model1.summary()
model1.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
model1.fit(x_train,y_train,epochs=25,validation_split=1/5,shuffle=True)
```

(e) . **MODEL2**

```
model2 = models.Sequential()
model2.add(layers.Conv2D(32,(5,5),activation='relu',input_shape=(32,32,3)))
model2.add(layers.MaxPool2D((3,3)))
model2.add(layers.Conv2D(64,(5,5),activation='relu'))
model2.add(layers.MaxPool2D((3,3)))
model2.add(layers.Flatten())
model2.add(layers.Dense(64,activation='relu'))
model2.add(layers.Dense(10,activation='softmax'))
model2.summary()
model2.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
model2.fit(x_train,y_train,epochs=25,validation_split=1/5,shuffle=True)
```

(f) . **MODEL3**

```
model3 = models.Sequential()
model3.add(layers.Conv2D(32,(3,3),activation='relu',input_shape=(32,32,3)))
model3.add(layers.MaxPool2D((2,2)))
model3.add(layers.Conv2D(64,(3,3),activation='relu'))
model3.add(layers.MaxPool2D((2,2)))
model3.add(layers.Flatten())
model3.add(layers.Dense(64,activation='relu'))
model3.add(layers.Dense(10,activation='softmax'))
model3.summary()
model3.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
model3.fit(x_train,y_train,epochs=25,validation_split=1/5,shuffle=True)
```

(g) . **MODEL4**

```
model4 = models.Sequential()
model4.add(layers.Conv2D(32,(3,3),activation='relu',input_shape=(32,32,3)))
model4.add(layers.MaxPool2D((2,2)))
model4.add(layers.Conv2D(64,(3,3),activation='relu'))
model4.add(layers.MaxPool2D((2,2)))
model4.add(layers.Conv2D(128,(3,3),activation='relu'))
model4.add(layers.MaxPool2D((2,2)))
model4.add(layers.Flatten())
model4.add(layers.Dense(64,activation='relu'))
model4.add(layers.Dense(10,activation='softmax'))
model4.summary()
model4.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
model4.fit(x_train,y_train,epochs=25,validation_split=1/5,shuffle=True)
```

(h) . MODEL4 is best model among the above models as it has high accuracy on test dataset compared to other models.

```
(loss1,accuracy1) = model1.evaluate(x_test,y_test)
print("model1_loss:",loss1,"model1_accuracy:",accuracy1)
(loss2,accuracy2) = model2.evaluate(x_test,y_test)
print("model2_loss:",loss2,"model2_accuracy:",accuracy2)
(loss3,accuracy3) = model3.evaluate(x_test,y_test)
print("model3_loss:",loss3,"model3_accuracy:",accuracy3)
(loss4,accuracy4) = model4.evaluate(x_test,y_test)
print("model4_loss:",loss4,"model4_accuracy:",accuracy4)
```

(i). **MODEL5**

```
#As the model4 is not possible to stride due to negative dimension model3 is used.
model5 = models.Sequential()
model5.add(layers.Conv2D(32,(3,3),strides=(2,2),activation='relu',input_shape=(32,32,3)))
model5.add(layers.MaxPool2D((2,2)))
model5.add(layers.Conv2D(64,(3,3),strides=(2,2),activation='relu'))
model5.add(layers.MaxPool2D((2,2)))
model5.add(layers.Flatten())
model5.add(layers.Dense(64,activation='relu'))
model5.add(layers.Dense(10,activation='softmax'))
model5.summary()
model5.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
model5.fit(x_train,y_train,epochs=25,validation_split=1/5,shuffle=True)
```

(j). **MODEL6**

```
model6 = models.Sequential()
model6.add(layers.Conv2D(32,(3,3),padding='same',activation='relu',input_shape=(32,32,3)))
model6.add(layers.MaxPool2D((2,2)))
model6.add(layers.Conv2D(64,(3,3),padding='same',activation='relu'))
model6.add(layers.MaxPool2D((2,2)))
model6.add(layers.Conv2D(128,(3,3),padding='same',activation='relu'))
model6.add(layers.MaxPool2D((2,2)))
model6.add(layers.Flatten())
model6.add(layers.Dense(64,activation='relu'))
model6.add(layers.Dense(10,activation='softmax'))
model6.summary()
model6.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
model6.fit(x_train,y_train,epochs=25,validation_split=1/5,shuffle=True)
```

(k). **MODEL7**

```
#As adding strides in model4 is not possible because of negative dimension we are using model3.
model7 = models.Sequential()
model7.add(layers.Conv2D(32,(3,3),strides=(2,2),padding='same',activation='relu',input_shape=(32,32,3)))
model7.add(layers.MaxPool2D((2,2)))
model7.add(layers.Conv2D(64,(3,3),strides=(2,2),padding='same',activation='relu'))
model7.add(layers.MaxPool2D((2,2)))
model7.add(layers.Flatten())
model7.add(layers.Dense(64,activation='relu'))
model7.add(layers.Dense(10,activation='softmax'))
model7.summary()
model7.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
model7.fit(x_train,y_train,epochs=25,validation_split=1/5,shuffle=True)
```

(l). model6 is the best model as it high test accuracy compared to others.

```
(loss5,accuracy5) = model5.evaluate(x_test,y_test)
print("model5_loss:",loss5,"model5_accuracy:",accuracy5)
(loss6,accuracy6) = model6.evaluate(x_test,y_test)
print("model6_loss:",loss6,"model6_accuracy:",accuracy6)
(loss7,accuracy7) = model7.evaluate(x_test,y_test)
print("model7_loss:",loss7,"model7_accuracy:",accuracy7)
```

(m). **MODEL8**

```
model8 = models.Sequential()
model8.add(layers.Conv2D(32,(3,3),padding='same',activation='tanh',input_shape=(32,32,3)))
model8.add(layers.MaxPool2D((2,2)))
model8.add(layers.Conv2D(64,(3,3),padding='same',activation='tanh'))
model8.add(layers.MaxPool2D((2,2)))
model8.add(layers.Conv2D(128,(3,3),padding='same',activation='tanh'))
model8.add(layers.MaxPool2D((2,2)))
model8.add(layers.Flatten())
model8.add(layers.Dense(64,activation='tanh'))
model8.add(layers.Dense(10,activation='softmax'))
model8.summary()
model8.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
model8.fit(x_train,y_train,epochs=25,validation_split=1/5,shuffle=True)
```

(n). **MODEL9**

```
model9 = models.Sequential()
model9.add(layers.Conv2D(32,(3,3),padding='same',activation='sigmoid',input_shape=(32,32,3)))
model9.add(layers.MaxPool2D((2,2)))
model9.add(layers.Conv2D(64,(3,3),padding='same',activation='sigmoid'))
model9.add(layers.MaxPool2D((2,2)))
model9.add(layers.Conv2D(128,(3,3),padding='same',activation='sigmoid'))
model9.add(layers.MaxPool2D((2,2)))
model9.add(layers.Flatten())
model9.add(layers.Dense(64,activation='sigmoid'))
model9.add(layers.Dense(10,activation='softmax'))
model9.summary()
model9.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
model9.fit(x_train,y_train,epochs=25,validation_split=1/5,shuffle=True)
```

```
(loss8,accuracy8) = model8.evaluate(x_test,y_test)
print("model8_loss:",loss8,"model8_accuracy:",accuracy8)
(loss9,accuracy9) = model9.evaluate(x_test,y_test)
print("model9_loss:",loss9,"model9_accuracy:",accuracy9)
```
